#### What is Machine Learning?

Machine learning is a method of data analysis that **automates** analytical model building.

Using algorithms (following a set of rules) that iteratively learn from data, machine learning allows computers to find hidden insights without being explicitly programmed on where to look.

ML is used for various tasks such as fraud detection, new price models, email spam filtering, ...

Neural networks are a specific type of machine learning process (mimicking biological neurons). These networks can be used to solve tasks that many other types of algorithms can not (Example: image classification).

#### Supervised Learning

Supervised learning algorithms are trained on labeled examples, such as inputs for which the desired output is known. Some historical features and historical labels are known to us.

Example: A segment of text could have a category label, such as spam vs legitimate email / postive vs negative movie review.

In neural networks (NN), based on inputs and desired outputs, the algorithm learns by comparing its output and the correct output. Then modify the model accordingly to minimize the error.

ML process:

Data Acquisition -> Data Cleaning -> Split Data (Train and Test) -> Build model using train data -> Test Model (update) -> Model Deployment

**Is it fair to use a single split?**

To address this, data is often split into 3 sets:

- Training Data (adjust model parameters)
- Validation Data (adjust model hyperparameters)
- Test Data (get final performance metric(s)/true performance). No adjustment to the model after this stage.


#### Overfitting and Underfitting

Overfitting fits too much to the noise in the data. Result in low error on the training set and high error in the test/validation sets.

Underfitting: The model does not capture the underlying trend of the data and does not fit the data well enough. Low variance but high bias. This often results of a excessively simple model.

When  we are dealing with multidimensional data, we would not be able to see the performance of the model visually. But we expect the error to be reduced when training time increases / epochs increase. However, for a bad model error may increase with epochs. 


#### Evaluating Performance - Classification Error Metrics

Key classification metrics we need to understand are:

- Accuracy
- Recall
- Precision
- F1-Score

Let's check on the reasoning behind these metrics and how they actually work.

In any classification task model can achieve two results: the model was correct OR model was incorrect.

Note that incorrect vs correct expands to situations where you have multiple classes (for understanding purposes, we can use binary classification).

Example: We attempt to predict if an image is a dog or a cat.

Since this is supervised learning, we will first fit/train a model on training data, and then test the model on testing data. Once we have model predictions, we will compare them to the true values (correct labels).

Process: We have a trained model, and we predict for different instances. At the end, we will have a count of correct matches and incorrect matches. 

**Note** that in the real world, not all incorrect or correct matches hold equal values (because it does not tell the full story).

Before looking into the metrics, we could organize the predicted values compared to the real values, which is known as a confusion matrix (Kind of a 2 by 2 matrix).

**Accuracy:**

- Number of correct predictions made by the model divided by the total number of predictions (correct predictions as a percentage).

Accuracy is useful when target classes are well-balanced. In our example, the number of cat images and dog images is roughly the same.

However, for unbalanced classes, accuracy is not a good choice. Example: 99 dog images and 1 cat image. If our model were simply a line that always predicts dog, we would have 99% accuracy.

**Recall:**

- Number of true positives divided by the number of true positives plus the number of false negatives.

That is, the ability of a model to find all the relevant cases within a dataset.

**Precision:**

- Number of true postoves devide by the number of true positives plus the number of false positives.

That is, the ability ofthe  classification model to identify only the relevant data points.

Recall and Precision: Often, we have a trade-off. While recall expresses the ability to find all relevant instances in a dataset, precision expresses the proportion of the data points our model says was relevant actually were relevant.

**F1-Score:**

This can be used to find an optimal blend of precision and recall (combine two metrics / harmonic mean).

$$F_{1} = 2*\frac{\text{precision}*\text{recall}}{\text{precision}+\text{recall}}$$

The reason for using the harmonic mean is that it punishes extreme values. Example: if a classifier with a precision of 1 and a recall of 0 has a simple average of 0.5. But an F1 score of 0.

**Model Evaluation:**

It is important to remember that a good metric really depends on the specific situation and the context.
Link: https://en.wikipedia.org/wiki/Confusion_matrix

Due to the precision/recall trade-off, we need to decide if the model should focus on fixing False Positives vs. False Negatives. In disease diagnosis, it is probably better go in the direction of False Positives. 


#### Evaluating Performance - Regression Error Metrics

Unlike in classification problems, regression models attempt to predict **continuous values**.

Popular metrics are:

**Mean Absolute Error (MAE):**

$$\frac{1}{n} \sum|y - \hat{y}|$$

The issue is that it does not punish large errors. We can address this with mean square errors.

**Mean Squared Errors (MSE):**

$$\frac{1}{n}\sum (y - \hat{y})^{2}$$

The issue is that it is squaring the units, and thus it is difficult to interpret. We can fix this with root mean square error (RMSE). The accuracy of the model (value of RMSE) can be verified with expert knowledge.
