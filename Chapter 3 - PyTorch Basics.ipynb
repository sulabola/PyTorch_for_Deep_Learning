### PyTorch Basics

Here will cover PyTorch tensors with some basics, how to create tensors, their operations, along with some exercises.

What is a tensor? A tensor is often thought of as a generalized matrix (1-D, 2-D, 3-D, ..., N-D tensors). It is a more generalized term.

Commaly used names:
- scalar
- 1-D == vector
- 2-D == Matrix
- 3-D and above == tensor

We use these tensors to arrange our data. When working with image data, it will be easiest to work with it as 3D tensors (view as different layers).

**Creating a tesnor from scratch in PyTorch**

```python
### Before moving forward, let's quickly check the difference between "tensor" and "Tensor".
### The main difference is data type.

new_arr = np.array([1,2,3])
new_arr
```

```python
new_arr.dtype
```

```python
torch.tensor(new_arr)
```

```python
torch.Tensor(new_arr)
```

```python
my_Tensor = torch.Tensor(new_arr)
my_Tensor.dtype
```
