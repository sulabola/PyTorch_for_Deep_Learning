### PyTorch Basics

Here will cover PyTorch tensors with some basics, how to create tensors, their operations, along with some exercises.

What is a tensor? A tensor is often thought of as a generalized matrix (1-D, 2-D, 3-D, ..., N-D tensors). It is a more generalized term.

Commaly used names:
- scalar
- 1-D == vector
- 2-D == Matrix
- 3-D and above == tensor

We use these tensors to arrange our data. When working with image data, it will be easiest to work with it as 3D tensors (view as different layers).

**Tensor Operations**

```python
### Create a 2D tensor

x = torch.arange(6).reshape(3,2)
x
```

```python
### Pick a single value from x

x[1,1]
```

```python
### Note: it returns as a tensor object
```

```python
type(x[1,1])
```

```python
### SLicers

x[:,1]
```

```python
### Note: Even though we are trying to get column 1, it does not retain as a column (we expect a 3 by 1 vector)/vertical vector.
### Instead, we see a 1 by 3 vector.
```

```python
### One way to solve this:

x[:,1:]
```

```python
x = torch.arange(10)
x
```

```python
### Or we can use "reshape":

x.reshape(2,5)
```

```python
### Let's assume we want to do: a = a + b
### We can perform this with "add". There is also a function "add_" (underscore method)
### Let's see how they are different (This works for other operations such as multiplication)

a.mul(b)
```

```python
a

### a is unchanged
```

```python
a.mul_(b)
```

```python
a

### a is replaced now.
```
